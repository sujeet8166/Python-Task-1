Q1. Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
Note: you must extract numeric keys and values of the dictionary also.

```python
def flatten_and_product(lst):
    flat_list = []

    def extract(item):
        if isinstance(item, (list, tuple, set)):
            for i in item:
                extract(i)
        elif isinstance(item, dict):
            for k, v in item.items():
                if isinstance(k, (int, float)) and not isinstance(k, bool):
                    flat_list.append(k)
                extract(v)
        elif isinstance(item, (int, float)) and not isinstance(item, bool):
            flat_list.append(item)
        # ignore strings and booleans

    extract(lst)

    product = 1
    for num in flat_list:
        product *= num
    return product

list1 = [1,2,3,4, [44,55,66, True], False, 
         (34,56,78,89,34), {1,2,3,3,2,1}, 
         {1:34, "key2":[55,67,78,89], 4:(45,22,61,34)}, 
         [56, 'data science'], 'Machine Learning']

print(flatten_and_product(list1))
```

Explanation: The function recursively extracts numbers from lists, tuples, sets, and dictionaries, Numeric keys and values from dictionaries are included, Non‑numeric values (strings, booleans) are ignored, Finally, the product of all numbers is calculated.

Q2. Write a python program for encrypting a message sent to you by your friend. The logic of encryption
should be such that, for a the output should be z. For b, the output should be y. For c, the output should
be x respectively. Also, the whitespace should be replaced with a dollar sign. Keep the punctuation
marks unchanged.
Input Sentence: I want to become a Data Scientist.
Encrypt the above input sentence using the program you just created.
Note: Convert the given input sentence into lowercase before encrypting. The final output should be
lowercase.

```python
def encrypt_message(msg):
    msg = msg.lower()
    result = ""
    for ch in msg:
        if ch.isalpha():
            result += chr(122 - (ord(ch) - 97))  # reverse alphabet
        elif ch == " ":
            result += "$"
        else:
            result += ch
    return result

sentence = "I want to become a Data Scientist."
encrypted = encrypt_message(sentence)
print(encrypted)
```

Explanation: Convert the input sentence to lowercase, Each alphabet character is replaced with its reverse (a→z, b→y, c→x, etc.), Spaces are replaced with $, Punctuation marks remain unchanged.
